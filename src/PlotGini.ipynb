```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
metrics = pd.read_csv("metrics.csv")
```

```python
metrics.describe(include="all")
```

```python
for feature, feature_metric in metrics.groupby("feature"):
    feature_metric.plot(x="split",y="metrics",title=feature,ylabel="G")
    plt.savefig(f"../doc/images/{feature}_plot.png")
```
